Q25. Write a Pandas program to 
a. filter all columns where all entries present, 
b. check which rows and columns has a NaN and 
c. finally drop rows with any NaNs from world alcohol consumption dataset.

```python
# Find  all columns which all entries present:
w_a_con.loc[:, w_a_con.notnull().all()]
```

```python
# ** Rows and columns has a NaN: (filter by rows)
w_a_con.loc[w_a_con.isnull().sum(axis=1)>0,:]
```

```python
# ** Rows and columns has a NaN: (filter by columns)
w_a_con.loc[:,w_a_con.isnull().any()]
```

```python
# Drop rows with any NaNs:")
w_a_con.dropna(how='any')
```
